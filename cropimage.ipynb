```python
# Get started

import cv2
from IPython.display import Image
from cropimage import Cropper
```

```python
# Prepare your dataset

# Randomly download an image from Unsplash

input = 'input.jpg'
!wget -q -O {input} 'https://unsplash.it/1920/1080/?random'
Image(input)
```

```python
cropper = Cropper()
result = cropper.crop(input)
```

```python
out = 'output.jpg'
cv2.imwrite(out, result)
Image(out)
```
